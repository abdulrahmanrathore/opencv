```python
# Import OpenCV library
import cv2

# Create a webcam object
cam = cv2.VideoCapture(0)

# Create a smile detector object
smile_detector = cv2.CascadeClassifier(cv2.data.haarcascades +"haarcascade_smile.xml")

# Create a loop that reads frames from webcam
while True:
    # Get a frame from webcam
    ret, frame = cam.read()

    # Check if frame was captured successfully
    if not ret:
        print("Failed to capture frame")
        break

    # Convert frame to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Detect smiles in frame
    smiles = smile_detector.detectMultiScale(gray, 1.3, 5)

    # Draw rectangles around smiles
    for (x, y, w, h) in smiles:
        cv2.rectangle(frame, (x, y), (x+w, y+h), (0, 255, 0), 2)

    # Check if there is at least one smile
    if len(smiles) > 0:
        # Save image
        cv2.imwrite("smile.png", frame)
        print("Image saved")

    # Show frame on screen
    cv2.imshow("Window", frame)

    # Check if user pressed ESC key
    if cv2.waitKey(1) & 0xFF == 27:
        # Exit loop
        break

# Release resources
cam.release()
cv2.destroyAllWindows()
```
